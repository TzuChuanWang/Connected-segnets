```python
from PIL import Image

def make_square(im, min_size=256, fill_color=(0, 0, 0, 0)):
    x, y = im.size
    size = max(min_size, x, y)
    new_im = Image.new('RGBA', (size, size), fill_color)
    new_im.paste(im, (int((size - x) / 2), int((size - y) / 2)))
    return new_im
```

# 隨機取照片

```python
import os, random, shutil
def copyFile(fileDir):
        pathDir = os.listdir(fileDir)    #取圖片的原始路徑
        filenumber=len(pathDir)
        rate=0.76    #自定義抽取圖片的比例，比方說100張抽10張，那就是0.1
        picknumber=int(filenumber*rate) #按照rate比例從資料夾中取一定數量圖片
        sample = random.sample(pathDir, picknumber)  #隨機選取picknumber數量的樣本圖片
        print (sample)
        for name in sample:
                shutil.copy(fileDir+name, tarDir+name)
        return

if __name__ == '__main__':
    fileDir = "E:/Alldata/hospital/croponly/image512/"    #源圖片資料夾路徑
    tarDir = "E:/Alldata/hospital/croponly/train/imagefortrain90/"    #移動到新的資料夾路徑
    copyFile(fileDir)
```
